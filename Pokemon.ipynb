```python
class Pokemon:
    def __init__(self, name, level, type):
        self.name = name
        self.level = level
        self.type = type
        self.max_health = level * 100
        self.current_health = level * 100
        self.knocked_out = False
    
    def __repr__(self):
        return "This level {level} {name} has {health} hit points remaining. They are a {type} type Pokemon".format(level = self.level, name = self.name, health=self.current_health, type = self.type)
    
    def lose_health(self, number):
        if self.current_health < number:
            self.current_health = 0
        else:
            self.current_health -= number
        if self.current_health == 0:
            self.knocked_out = True
        return "{name} lost {number} health points, now has {current_health}.".format(name = self.name, number = number, current_health = self.current_health)
    
    def regain_health(self, number):
        self.current_health += number
        if self.current_health > self.max_health:
            self.current_health = self.max_health
        if self.current_health > 0:
            self.knocked_out = False
        return "{name} regain {number} health points, now has {current_health}.".format(name = self.name, number = number, current_health = self.current_health)
    
    def attack(self, opponent):
        if self.knocked_out:
            return 'Your {name} is knocked out.'.format(name = self.name)
        else:
            number = self.level * 10
            if self.type == 'Fire' and opponent.type == 'Water':
                force_attack = 0.5 * number
            elif self.type == 'Fire' and opponent.type == 'Grass':
                force_attack = 2 * number
            elif self.type == 'Water' and opponent.type == 'Grass':
                force_attack = 0.5 * number
            elif self.type == 'Water' and opponent.type == 'Fire':
                force_attack = 2 * number
            elif self.type == 'Grass' and opponent.type == 'Fire':
                force_attack = 0.5 * number
            elif self.type == 'Grass' and opponent.type == 'Water':
                force_attack = 2 * number
            opponent.lose_health(force_attack)
            return '{opponent_name} was attacked by {name} and lost {number} health points, now has {opponent_current_health}'.format(opponent_name = opponent.name, name = self.name, number = force_attack, opponent_current_health = opponent.current_health)
        
```

```python
class Trainer:
    def __init__(self, pokemons, name, num_potions, active_pokemon):
        self.pokemons = pokemons
        self.name = name
        self.num_potions = num_potions
        self.active_pokemon = active_pokemon
    def use_potion(self):
        if self.num_potions == 0:
            return 'Out of potions'
        else:
            self.num_potions -= 1
            self.pokemons[self.active_pokemon].regain_health(100)
            return '+100 health points, {name} now has {curent_health}'.format(name = self.pokemons[self.active_pokemon].name, curent_health = self.pokemons[self.active_pokemon].current_health)
    def attack(self, opponent):
        return self.pokemons[self.active_pokemon].attack(opponent.pokemons[opponent.active_pokemon])
    def switch_pokemon(self, number):
        if number >= 0 and number < len(self.pokemons) and self.pokemons[self.active_pokemon].knocked_out == False:
            self.active_pokemon = number
            return 'You chose {name}.'.format(name = self.pokemons[self.active_pokemon].name)
        else:
            return "Can't do that"
```
